# Game `defeat the monster`

### This game will be a turn-based game in which a player will choose a weapon and will fight a bear in order to survive.

### The player has an amount of life and stamina and will hit the monster each turn. The amount of damage will take in account the stamina left and the weapon of choice. Each hit will take stamina out of you. In the Monster's turn, the amount of damage will be determined by the stamina left and the presence or not of a shield. There will be different items offered after each round that will affect the stamina\life\critical hit chance of the player

## lets begin bulding! I would need a number of functions that will determine all of the above, running after each turn.

```python
def show_health(health):
    maxHealth = 400
    if health < 0:
        health = 0
    if health > maxHealth:
        health = maxHealth
    healthDashes = 40
    dashConvert = int(maxHealth/healthDashes)            # Get the number to divide by to convert health to dashes (being 10)
    currentDashes = int(health/dashConvert)              # Convert health to dash count: 80/10 => 8 dashes
    remainingHealth = healthDashes - currentDashes       # Get the health remaining to fill as space => 12 spaces

    healthDisplay = '=' * currentDashes                  # Convert 8 to 8 dashes as a string:   "--------"
    remainingDisplay = ' ' * remainingHealth             # Convert 12 to 12 spaces as a string: "            "
    percent = str(int((health/maxHealth)*100)) + "%"     # Get the percent as a whole number:   40%

    print("                          |" + healthDisplay + remainingDisplay + "|")  # Print out textbased healthbar
    print("                           " + (str(maxHealth)+" / "+str(health)))                         # Print the percent
```

```python
def show_stamina(stamina):
    maxStamina = 50
    if stamina > 50:
        stamina = 50
    if stamina < 0:
        stamina = 0
    staminaDashes = 20
    dashConvert = int(maxStamina/staminaDashes)            # Get the number to divide by to convert health to dashes (being 10)
    currentDashes = int(stamina/dashConvert)              # Convert health to dash count: 80/10 => 8 dashes
    remainingStamina = staminaDashes - currentDashes       # Get the health remaining to fill as space => 12 spaces

    staminaDisplay = '=' * currentDashes                  # Convert 8 to 8 dashes as a string:   "--------"
    remainingDisplay = ' ' * remainingStamina             # Convert 12 to 12 spaces as a string: "            "
    percent = str(int((stamina/maxStamina)*100)) + "%"     # Get the percent as a whole number:   40%

    print("                          |" + staminaDisplay + remainingDisplay + "|")  # Print out textbased healthbar
    print("                           " + (str(maxStamina)+" / "+str(stamina)))                         # Print the percent
```

```python
import random
def player_attack(weapon, stamina):
    """calculates the damage from the player"""
    if stamina > 0:
        damage = int((weapon * (0.9* stamina) * random.randint(1,10))/10)
        
        return damage
    else:
        print("You have no stamina left! missed attack!")
        return damage
```

```python
def player_choice():
    """makes the user input 1 2 3 for the 3 available actions"""
    option = 0
    
    while option not in list(range(1,4)):
        option = int(input("          Choose your next action!: 1 for attack | 2 to use defense | 3 to use an item "))
    
    return option
```

```python
## Start of the game

start = intro().lower()
if start[0] == "y":
    game_on = True
else:
    game_on = False

    #Initialise stats 

weapon=10
health = 400
stamina = 50
monster_health = 1000
turn = "player"
poison = False
stun = False
    #game starts with player turn
while game_on:
    
    if turn == "player":
        if health <= 0:
            game_on = False
        else:
            #player chooses action
            clear_output()
            show_player_screen(health, stamina)
            option = player_choice()
            
            
            # Attack logic. you deal an amount of damage from the player attack function. 
            #If the target is stun you double the damage for this turn. Stun turns back to False after 1 turn
            if option == 1:
                if stun:
                    damage = (player_attack(weapon,stamina) * 1.5) #50% more damage to a stunned target
                    stun = False
                    clear_output()
                    animate_weapon()
                    print(f"                                CRITICAL HIT! Attacked the monster for {damage} points!")
                    monster_health -= damage
                    stamina -= 15
                    defense = False
                    #poison logic
                    if poison:
                        health -= 5
                        print("                               You loose 5 HP due to poison damage")
                        
                    
                    time.sleep(2)
                    clear_output()
                    show_weapon()
                    next_step()
                else:
                    
                    damage = player_attack(weapon, stamina) #you loose stamina after each attack
                    clear_output()
                    animate_weapon()
                    
                    print(f"                               Attacked the monster for {damage} points!")
                    monster_health -= damage
                    stamina -= 15
                    defense = False
                    #poison logic
                    if poison:
                        health -= 5
                        print("                              You loose 5 HP due to poison damage")
                    time.sleep(2)
                    clear_output()
                    
                    next_step()
            
            #DEFEND action. if you choose this the variable defense becomes true and that affects the damage function
            #of the enemy. Also there is a high possibility that you stun the enemy. If this happens the variable
            #stun is activated which will affect the damage player function
            elif option == 2:
                stamina += 2 
                clear_output()
                show_shield()
                print(f"                         The warrior defends against the next attack!")
                defense = True
                if poison:
                        health -= 5
                        print("                      You loose 5 HP due to poison damage")
                if random.randint(1,20)>6:
                    stun = True
                time.sleep(2)
                clear_output()
                next_step()
                
            #Consume object action. you can choose a health potion or stamina, which will increase HP\stamina. takes 1
            #turn. A potion eliminates poison effect
            
            elif option == 3:
                stamina += 10 
                if poison:
                        health -= 5
                        print("                    You loose 5 HP due to poison damage")
                        
                selection = use_item()
                if selection == 1:
                    health = use_health_potion(health)
                    clear_output()
                    show_potion()
                    time.sleep(1)
                    clear_output()
                    show_player_screen(health,stamina)
                    print("                         You healed for 30 HP")
                    if poison:
                        print("                    The potion cured the poison!")
                        poison = False
                        
                    time.sleep(2)
                    clear_output()
                    next_step()
                elif selection == 2:
                    stamina = use_stamina_potion(stamina)
                    clear_output()
                    show_potion()
                    time.sleep(1)
                    clear_output()
                    show_player_screen(health, stamina)
                    print("                          You recovered 30 stamina")
                    if poison:
                        print("                     The potion cured the poison!")
                        poison = False
                       
                    time.sleep(2)
                    clear_output()
                    next_step()
            
            #after one of the actions are completed turn goes to the monster
            turn = "monster"
    else:
        if monster_health <= 0:
            game_on = False
        
        else:
        
            #stun logic. Monster misses the turn if stun. Stun goes back to false after the turn
            
            if stun:
                clear_output()
                show_monster_screen(monster_health)
                print(f"                                   The monster is stun! could not attack you!")
                stamina += 2 
                time.sleep(2)
                clear_output()
                next_step()
                turn = "player"
                
            else:
                clear_output()
                show_monster_screen(monster_health)
                time.sleep(2)
                clear_output()
                animate_spider()
                monster_damage = monster_attack(defense)
                print(f"                                   The monster attacked for {monster_damage} points! ")
                stamina += 2 
                health -= monster_damage
                
                
                #poison logic. During monster attack poison can be activated. This will take hp each turn. Takes
                #a health potion to eliminate effect
                if random.randint(1,20)<7:
                    poison = True
                if poison:
                    print("                               The monster has infected you with its poison!")
                    
                time.sleep(2)
                clear_output()
                next_step()
                turn = "player"

if health > monster_health:
    clear_output()
    animate_player_intro()
    print("                         Congratulations! you defeated the Elder Spider!")
else:
    clear_output()
    animate_dead_player()
    print("                         The monster defeated you! better luck next time!")
```
