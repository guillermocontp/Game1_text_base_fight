# Game `defeat the monster`

This game will be a turn-based game in which a player will choose a weapon and will fight a bear in order to survive.

The player has an amount of life and stamina and will hit the monster each turn. The amount of damage will take in account the stamina left and the weapon of choice. Each hit will take stamina out of you. In the Monster's turn, the amount of damage will be determined by the stamina left and the presence or not of a shield. There will be different items offered after each round that will affect the stamina\life\critical hit chance of the player

lets begin bulding! I would need a number of functions that will determine all of the above, running after each turn.

let's start with the character, define it and its weapons

### This is a simple text based game made in basic python. In this iteration I will now write whatever I can in classes. A next iteration will use pygame for whatever I can improve

```python
class Player:
    
    def __init__(self, health, stamina, weapon):
        self.health = health
        self.weapon = weapon
        self.stamina = stamina
    
    def show_health(self):
        maxHealth = 300 
        if self.health < 0:
            self.health = 0
        if self.health > maxHealth:
            self.health = maxHealth
        healthDashes = 40
        dashConvert = int(maxHealth/healthDashes)            # Get the number to divide by to convert health to dashes (being 10)
        currentDashes = int(self.health/dashConvert)              # Convert health to dash count: 80/10 => 8 dashes
        remainingHealth = healthDashes - currentDashes       # Get the health remaining to fill as space => 12 spaces

        healthDisplay = '=' * currentDashes                  # Convert 8 to 8 dashes as a string:   "--------"
        remainingDisplay = ' ' * remainingHealth             # Convert 12 to 12 spaces as a string: "            "
        percent = str(int((self.health/maxHealth)*100)) + "%"     # Get the percent as a whole number:   40%

        print("                          |" + healthDisplay + remainingDisplay + "|")  # Print out textbased healthbar
        print("                           " + (str(self.health)+" / "+str(maxHealth)))                         # Print the percent
        
    
    def show_stamina(self):
        maxStamina = 50
        if self.stamina > maxStamina:
            self.stamina = maxStamina
        if self.stamina < 0:
            self.stamina = 0
        staminaDashes = 20
        dashConvert = int(maxStamina/staminaDashes)            # Get the number to divide by to convert health to dashes (being 10)
        currentDashes = int(self.stamina/dashConvert)              # Convert health to dash count: 80/10 => 8 dashes
        remainingStamina = staminaDashes - currentDashes       # Get the health remaining to fill as space => 12 spaces

        staminaDisplay = '=' * currentDashes                  # Convert 8 to 8 dashes as a string:   "--------"
        remainingDisplay = ' ' * remainingStamina             # Convert 12 to 12 spaces as a string: "            "
        percent = str(int((self.stamina/maxStamina)*100)) + "%"     # Get the percent as a whole number:   40%

        print("                          |" + staminaDisplay + remainingDisplay + "|")  # Print out textbased healthbar
        print("                           " + (str(self.stamina)+" / "+str(maxStamina)))          # Print the percent
        
    
    
    def player_attack(self):
        """calculates the damage from the player"""
        if self.stamina > 0:
            damage = int((self.weapon * (0.9* stamina) * random.randint(1,10))/10)
    
        else:
            print("You have no stamina left! missed attack!")
            damage = 0
        return damage
    
    
    def player_choice(self):
        """makes the user input 1 2 3 for the 3 available actions"""
        option = 0

        while option not in list(range(1,4)):
            option = int(input("          Choose your next action!: 1 for attack | 2 to use defense | 3 to use an item "))

        return option


    def defend(self):
        """this will take in account the stamina left to come up with an amount of defense to add to the attack function 
        of the monster. it will also guarantee that you get no critical"""
        defense = self.stamina * random.randint(1,6)
            #critical= False lets leave this for the end
        return defense    
    

    
    
    def show_player(self):
        print(r"""   

                                                     .--.
                                                    /.--.\
                                                    |====|
                                                    |`::`|
                                                .-;`\..../`;-.
                                               /  |...::...|  \
                                              |   /'''::'''\   |
                                              ;--'\   ::   /\--;
                                              <__>,>._::_.<,<__>
                                              |  |/   ^^   \|  |
                                              \::/|        |\::/
                                              |||\|        |/|||
                                              ''' |___/\___| '''
                                                   \_ || _/
                                                   <_ >< _>
                                                   |  ||  |
                                                   |  ||  |
                                                  _\.:||:./_
                                            jgs  /____/\____\


              """)
    
    def show_dead_player_1(self):
        print(r"""   

                        |---.\                                   .--.
                ___     |    `                                  /.--.\
               / .-\  ./=)                                      |====|
              |  |"|_/\/|                                       |`::`|
              ;|-;| /_  |                                   .-;`\..../`;-.
             / \_| |/ \ |                                  /  |...::...|  \
            /      \/\( |                                 |   /'''::'''\   |
            |   /  |` ) |                                 ;--'\   ::   /\--;
            /   \ _/    |                                 <__>,>._::_.<,<__>
           /--._/  \    |                                 |  |/   ^^   \|  |
           `/|)    |    /                                 \::/|        |\::/ 
             /     |   |                                  |||\|        |/|||
           .'      |   |                                  ''' |___/\___| '''
          /         \  |                                       \_ || _/
         (_.-.__.__./  /                                       <_ >< _>
                                                               |  ||  |
                                                               |  ||  |
                                                              _\.:||:./_
                                                             /____/\____\         




              """)
    
    
    def show_dead_player_2(self):
    
        print(r"""   

                                |---.\                           .--.
                        ___     |    `                          /.--.\
                       / .-\  ./=)                              |====|
                      |  |"|_/\/|                               |`::`|
                      ;|-;| /_  |                           .-;`\..../`;-.
                     / \_| |/ \ |                          /  |...::...|  \
                    /      \/\( |                         |   /'''::'''\   |
                    |   /  |` ) |                         ;--'\   ::   /\--;
                    /   \ _/    |                         <__>,>._::_.<,<__>
                   /--._/  \    |                         |  |/   ^^   \|  |
                   `/|)    |    /                         \::/|        |\::/ 
                     /     |   |                          |||\|        |/|||
                   .'      |   |                          ''' |___/\___| '''
                  /         \  |                               \_ || _/
                 (_.-.__.__./  /                               <_ >< _>
                                                               |  ||  |
                                                               |  ||  |
                                                              _\.:||:./_
                                                             /____/\____\         




              """)
    
    
    def show_dead_player_3(self):
        print(r"""   
    
                                            |---.\           .--.
                                    ___     |    `          /.--.\
                                   / .-\  ./=)              |====|
                                  |  |"|_/\/|               |`::`|
                                  ;|-;| /_  |           .-;`\..../`;-.
                                 / \_| |/ \ |          /  |...::...|  \
                                /      \/\( |         |   /'''::'''\   |
                                |   /  |` ) |         ;--'\   ::   /\--;
                                /   \ _/    |         <__>,>._::_.<,<__>
                               /--._/  \    |         |  |/   ^^   \|  |
                               `/|)    |    /         \::/|        |\::/ 
                                 /     |   |          |||\|        |/|||
                               .'      |   |          ''' |___/\___| '''
                              /         \  |               \_ || _/
                             (_.-.__.__./  /               <_ >< _>
                                                           |  ||  |
                                                           |  ||  |
                                                          _\.:||:./_
                                                         /____/\____\         

          """)
        
        
    def animate_dead_player(self):
        self.show_dead_player_1()
        time.sleep(0.7)
        clear_output()
        self.show_dead_player_2()
        time.sleep(0.7)
        clear_output()
        self.show_dead_player_3()
        self.show_health()
    
    
    def show_player_screen(self):
        """show player picture and health and stamina stats"""
        self.show_player()
        self.show_health()
        self.show_stamina()
        
        
    def show_weapon(self):
        print(r"""     


                                             /!\
                                            / ! \
                                     /\     )___(
                                    (  `.____(_)_________
                                    |           __..--""
                                    (       _.-|
                                     \    ,' | |
                                      \  /   | |
                                       \(    | |
                                        `    | |
                                             | |
                                             | | 
                                             | | 
                                             | | 
                                             | | 
                                             | | 
                                            (   )
                                            (   )
                                             --- 

                        """)
        
        
    def show_weapon_2(self):
        print(r"""     


                                                         /!\
                                                        / ! \
                                                 /\     )___(
                                                (  `.____(_)_________
                                                |           __..--""
                                                (       _.-|
                                                 \    ,' | |
                                                  \  /   | |
                                                   \(    | |
                                                    `    | |
                                                         | |
                                                         | | 
                                                         | | 
                                                         | | 
                                                         | | 
                                                         | | 
                                                        (   )
                                                        (   )
                                                         --- 

        """)
        
    def animate_weapon(self):
        self.show_weapon()
        time.sleep(0.7)
        clear_output()
        self.show_weapon_2()
        time.sleep(0.7)
        clear_output()
        self.show_weapon()

        time.sleep(0.7)
        clear_output()
        self.show_weapon_2()
        
        
    def show_shield(self):
        print(r"""   

                                                      .--.
                                                     /.--.\
                                                     |====|
                                                     |`::`|
                                                 .-;`\..../`;_.-^-._
                                                /  |...::..|`   :   `|
                                               |   /'''::''|   .:.   |
                                               ;--'\   ::  |..:::::..|
                                               <__> >._::_.| ':::::' |
                                               |  |/   ^^  |   ':'   |
                                               \::/|       \    :    /
                                               |||\|        \   :   /
                                               ''' |___/\___|`-.:.-`
                                                    \_ || _/    `
                                                    <_ >< _>
                                                    |  ||  |
                                                    |  ||  |
                                                   _\.:||:./_
                                             jgs  /____/\____\


        """)
```

```python
def use_item():
    """if the player chooses this"""
    selection = 0
    while selection not in list(range(1,3)):
        selection = int(input("           Select your item: 1 for health potion | 2 for stamina potion"))
    
    return selection
    
        
```

```python
def intro():
    """Funtion to start the intro story. shows the animated knight and asks for Y to start the game"""
    show_intro_title()
    print("                         THIS IS THE STORY OF THE KNIGHT THAT FOUGHT BACK THE ELDER SPIDER")
    time.sleep(2)
    clear_output()
    animate_player_intro()
    
    a = input("                              WILL YOU HELP HIM DEFEAT THE MONSTER?  Y/N")
    clear_output()
    monster.animate_spider()
    print("                                      READY YOUR WEAPONS!")
    next_step()
    return a
    
```

```python
def next_step():
    """Waits for an input to continue the action"""
    marker = ""
    while marker == "":
        input("press any key to continue: ")
        break
```

```python
def character_selection():
    health = int(input('Select the health of your character'))
    clear_output()               
    stamina = int(input('Select the stamina of your character'))
    clear_output()
    weapon = int(input('select weapon (damage)'))
    clear_output()
    return health, stamina, weapon
    
```

# Game logic

```python
## Start of the game
monster = Spider(2000, 20)
start = intro().lower()
if start[0] == "y":
    game_on = True
else:
    game_on = False

    #Choose character

health, stamina, weapon = character_selection()
player = Player(health, stamina, weapon)

turn = "player"
poison = False
stun = False
    #game starts with player turn
while game_on:
    
    if turn == "player":
        if (player.health <= 0) or (monster.monster_health <= 0):
            game_on = False
        else:
            #player chooses action
            clear_output()
            player.show_player_screen()
            option = player.player_choice()
            
            
            # Attack logic. you deal an amount of damage from the player attack function. 
            #If the target is stun you double the damage for this turn. Stun turns back to False after 1 turn
            if option == 1:
                if stun:
                    damage = (player.player_attack() * 2) #double damage for stun target
                    stun = False
                    clear_output()
                    player.animate_weapon()
                    print(f"                             CRITICAL HIT! Attacked the monster for {damage} points!")
                    monster.monster_health -= damage
                    player.stamina -= 15
                    defense = False
                    #poison logic
                    if poison:
                        player.health -= 5
                        print("you loose 5 HP due to poison damage")
                        
                    
                    time.sleep(2)
                    clear_output()
                    player.show_weapon()
                    next_step()
                else:
                    
                    damage = player.player_attack() #you loose stamina after each attack
                    clear_output()
                    player.animate_weapon()
                    
                    print(f"                               Attacked the monster for {damage} points!")
                    monster.monster_health -= damage
                    player.stamina -= 15
                    defense = False
                    #poison logic
                    if poison:
                        player.health -= 5
                        print("                         You loose 5 HP due to poison damage")
                    time.sleep(2)
                    clear_output()
                    
                    next_step()
            
            #DEFEND action. if you choose this the variable defense becomes true and that affects the damage function
            #of the enemy. Also there is a high possibility that you stun the enemy. If this happens the variable
            #stun is activated which will affect the damage player function
            elif option == 2:
                
                clear_output()
                player.show_shield()
                print(f"                         The warrior defends against the next attack!")
                defense = True
                if poison:
                        player.health -= 5
                        print("                   You loose 5 HP due to poison damage")
                if random.randint(1,20)>6:
                    stun = True
                time.sleep(2)
                clear_output()
                next_step()
                
            #Consume object action. you can choose a health potion or stamina, which will increase HP\stamina. takes 1
            #turn. A potion eliminates poison effect
            
            elif option == 3:
                
                if poison:
                        player.health -= 5
                        print("                    You loose 5 HP due to poison damage")
                        
                selection = use_item()
                if selection == 1:
                    player.health += 50
                    clear_output()
                    show_potion()
                    time.sleep(1)
                    clear_output()
                    player.show_player_screen()
                    print("                         You healed for 50 HP")
                    if poison:
                        print("                    The potion cured the poison!")
                        poison = False
                        
                    time.sleep(2)
                    clear_output()
                    next_step()
                elif selection == 2:
                    player.stamina =+ 50
                    clear_output()
                    show_potion()
                    time.sleep(1)
                    clear_output()
                    player.show_player_screen()
                    print("                          You recovered 30 stamina")
                    if poison:
                        print("                     The potion cured the poison!")
                        poison = False
                       
                    time.sleep(2)
                    clear_output()
                    next_step()
            
            #after one of the actions are completed turn goes to the monster
            turn = "monster"
    else:
        if (player.health <= 0) or (monster.monster_health <= 0):
            game_on = False
            
        else:
        
            #stun logic. Monster misses the turn if stun. Stun goes back to false after the turn
            
            if stun:
                clear_output()
                monster.show_monster_screen()
                print(f"                                   The monster is stun! could not attack you!")
                
                time.sleep(2)
                clear_output()
                next_step()
                turn = "player"
                
            else:
                clear_output()
                monster.show_monster_screen()
                time.sleep(2)
                clear_output()
                monster.animate_spider()
                monster_damage = monster.monster_attack(defense)
                
                print(f"                                   The monster attacked for {monster_damage} points! ")
                 
                player.health -= monster_damage
                
                
                #poison logic. During monster attack poison can be activated. This will take hp each turn. Takes
                #a health potion to eliminate effect
                if random.randint(1,20)<7:
                    poison = True
                if poison:
                    print("                               The monster has infected you with its poison!")
                    
                time.sleep(2)
                clear_output()
                next_step()
                turn = "player"

if monster.monster_health <= 0:
    clear_output()
    animate_player_intro()
    print("                         Congratulations! you defeated the Elder Spider!")
elif player.health <= 0:
    clear_output()
    player.animate_dead_player()
    print("                         The monster defeated you! better luck next time!")
else:
    print('something is wrong, this message should not appear')
```
